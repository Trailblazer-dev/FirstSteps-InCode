```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy  as np
import pandas as pd
```

```python
# fix random seed for reproducibility
seed =7
np.random.seed(seed)
```

## Load the Data
Pima Indian dataset

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
dataset = df.to_numpy()
```

```python
X = dataset[:,0:8]
Y = dataset[:,8]
```

## Create Model

```python
model = Sequential()
model.add(Dense(12,input_dim=8,kernel_initializer='uniform',activation='relu'))
model.add(Dense(8,kernel_initializer='uniform',activation='relu'))
model.add(Dense(1,kernel_initializer='uniform',activation='sigmoid'))
```

## Compile Model

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

## Fit Model
The training process will run for a fixed number of iterations through the dataset called
**epochs**

```python
model.fit(X,Y,epochs=150,batch_size=10)
```

## Evaluate Model
We have trained our neural network on the entire dataset and we can evaluate the performance
of the network on the same dataset


This will only give us an idea of how well we have modeled
the dataset (e.g. train accuracy), but no idea of how well the algorithm might perform on new
data.

```python
scores = model.evaluate(X,Y)
print("%s: %.2f%%" % (model.metrics_names[1],scores[1]*100))
```
